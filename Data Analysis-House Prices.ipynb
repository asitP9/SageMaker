## Histogram Plot

```python
plt.figure()
bostonHouseFrame.hist(figsize=(20,20))
plt.show()
```

## BOX PLOT

```python
bostonHouseFrame.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=False, figsize=(20,20))
plt.show()
```

```python
import seaborn as sns
correlations=bostonHouseFrame.corr()

# Set up the matplotlib figure
f, ax=plt.subplots(figsize=(11, 9))

#Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlations, cmap="PuBu", center=0, vmin=-1, vmax=1, square=True)

#fix for mpl bug that cuts off top/bottom of seaborn viz
b,t=plt.ylim() #Discover the values for bottom and top
b+=0.5
t-=0.5
plt.ylim(b,t)#update the ylim(bottom, top) values
plt.show()
```

## Scatter Plot


```python
plt.scatter(bostonHouseFrame['RM'], bostonHouseFrame['MEDV'])
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner occupied homes in $1000s')
plt.show()
```

```python
plt.scatter(bostonHouseFrame['LSTAT'], bostonHouseFrame['MEDV'])
plt.xlabel('% lower status of the population')
plt.ylabel('Median value of owner occupied homes in $1000s')
plt.show()
```
